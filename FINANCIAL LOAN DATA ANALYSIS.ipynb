# **FINANCIAL LOAN DATA ANALYSIS**

#### **Importing Libraries**

```python
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
```

####  **Load Data**



```python
df = pd.read_excel("financial_loan.xlsx")

df.head()
```

#### **Handling Missing Values**

```python
df.drop(columns =['emp_title'], inplace = True)

df.isnull().sum()
```

#### **Total Loan Applications**

```python
total_loan_applications = df['id'].nunique()

print("Total Loan Applications : ", total_loan_applications)
```

#### **Total Funded Amount**

```python
total_funded_amount = df['loan_amount'].nunique()

print("Total Funded Amount : ", total_funded_amount)
```

#### **Total Amount Recieved**

```python
total_amount_recieved = df['total_payment'].sum()

print("Total Amount Recieved : ", total_amount_recieved)
```

#### **Average Interest Rate**

```python
average_interest_rate = df['int_rate'].mean()

print("Average Interest Rate : ", average_interest_rate)
```

#### **Average Debt - to - Income (DTI)**

```python
average_dti = df['dti'].mean()

print("Average Debt-to-Income : " , average_dti)
```

#### **Good Loan VS Bad Loan**

```python
good_loan = df[df['loan_status'].isin(['Fully Paid', 'Current'])]

bad_loan = df[df['loan_status'] == 'Charged Off']


good_loan_percent = (len(good_loan) / len(df)) * 100

bad_loan_percent = (len(bad_loan) / len(df)) * 100

print("\nGood Loan Applications:", len(good_loan))

print("Bad Loan Applications:", len(bad_loan))

print(f"Good Loan %: {good_loan_percent:.2f}%")

print(f"Bad Loan %: {bad_loan_percent:.2f}%")


print("\nGood Loan Funded Amount:", good_loan['loan_amount'].sum())
print("Bad Loan Funded Amount:", bad_loan['loan_amount'].sum())
print("Good Loan Received Amount:", good_loan['total_payment'].sum())
print("Bad Loan Received Amount:", bad_loan['total_payment'].sum())
```

#### **Loan Status Distribution**

```python
plt.figure(figsize=(6,6))

df['loan_status'].value_counts().plot.pie(autopct = '%1.1f%%' , startangle = 90)

plt.title("LOAN STATUS DISTRIBUTION")

plt.ylabel(" ")

plt.show()
```

#### **Monthly Loan Applications Trend**

```python
df['Month'] = df['issue_date'].dt.to_period('M')

monthly_trend = df.groupby('Month')['id'].count()

plt.figure(figsize = (10,5))

monthly_trend.plot(kind = 'line' , marker = 'o', color = 'teal')

plt.title("MONTHLY LOAN APPLICATIONS TREND")

plt.xlabel("Month")

plt.ylabel("Number Of Applications")

plt.grid(True)

plt.show()
```

#### **Top 10 States by Loan Applications**

```python
top_states = df['address_state'].value_counts().head(10)

plt.figure(figsize = (10,5))

top_states.plot(kind = 'bar' , color = 'skyblue')

plt.title("TOP 10 STATES BY LOAN APPLICATIONS")

plt.xlabel("State")

plt.ylabel("Number Of Applications")

plt.show()
```

#### **Loan Term Distribution**

```python
plt.figure( figsize = (6,6))

df['term'].value_counts().plot.pie(autopct = '%1.1f%%' , startangle = 90)

plt.title("LOAN TERM DISTRIBUTION")

plt.ylabel(" ")

plt.show()
```

#### **Employment Length Analysis**

```python
plt.figure(figsize = (10,5))

df['emp_length'].value_counts().sort_index().plot(kind = 'bar' , color = 'orange')

plt.title("EMPLOYMENT LENGTH ANALYSIS")

plt.xlabel("Employment Length")

plt.ylabel("Applications")

plt.show()
```

#### **Loan Purpose Breakdown**

```python
plt.figure(figsize =(10,5))

df['purpose'].value_counts().plot(kind = 'bar' , color = 'green')

plt.title("LOAN PURPOSE BREAKDOWN")

plt.xlabel("Purpose")

plt.ylabel("Count")

plt.show()
```

#### **Home Ownership Analysis**

```python

df['home_ownership'] = df['home_ownership'].astype(str).str.strip().str.upper()


df['home_ownership'] = df['home_ownership'].replace({
    'NONE/OWN': 'OWN',
    'OTHER/OWN': 'OWN',
    'RENT/OWN': 'RENT',
    'MORTGAGE/OWN': 'MORTGAGE',
    'NONE': 'OTHER'
})


plt.figure(figsize=(6,6))
df['home_ownership'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("HOME OWNERSHIP DISTRIBUTION")
plt.ylabel("")
plt.show()

```
